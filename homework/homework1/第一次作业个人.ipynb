### 读入数据

```python
regional_file = path_data + "地区板块.txt" # 板块分析结果文件名，含路径
df_regionalCode = pd.read_csv(regional_file, encoding="gbk", sep="\t", dtype=str,names=["板块代码", "板块名称", "股票代码", "股票名称"])
df_regionalCode
```
