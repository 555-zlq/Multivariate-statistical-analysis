我们选取了加息时间跨度, 加息额度 + 美国GDP+ 加息次数+年通胀率+失业率作为自变量，选取GDP增长率和货币汇率变化作为因变量，选取中国等10个国家分析，通过对每个国家两个因变量分别进行多元线性回归的建模，最终得出加息时间跨度, 加息额度 + 美国GDP+ 加息次数在某些国家上具有明显的线性关系，而年通胀率+失业率影响都不显著，我们分别选取了南非 GDP 印度 GDP 哥伦比亚 货币汇率 南非 货币汇率四个模型进行展示

有趣的发现：瑞士，中国的各项指标都没有显著的关系，可能瑞士具有较稳定的经济体系
受影响较大的都是一些落后的不发达的国家

```python
file_path = path_data + "homework3.xlsx"
df_data = pd.read_excel(file_path, sheet_name="哥伦比亚货币") # 读取数据，指定第1列为索引
df_data.head() # 显示前5个记录，观察数据读取是否正确
```

### 通过绘制矩阵图和相关系数热力图,找出哪些变量与因变量存在线性关系

```python
sns.heatmap(df_data.corr(),annot=True,fmt=".2f", cmap="coolwarm",annot_kws={ "size":14},mask=(df_data.corr()==1))
```

```python
sns.pairplot(df_data, height=1.2, aspect=1.3,diag_kind="kde")
plt.tight_layout()
plt.show()
```

```python
model = sm.formula.ols(formula="货币汇率变化~ 加息时间跨度 + 加息额度 + 美国GDP+ 加息次数+年通胀率+失业率", data=df_data).fit()
print(model.summary())
```
